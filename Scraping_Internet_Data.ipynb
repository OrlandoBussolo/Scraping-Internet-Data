```python
# VERIFING WEBSITE PROPERTIES
import requests
from requests.exceptions import HTTPError

robots = None
URL = "https://www.imdb.com/robots.txt"

try:
  resposta = requests.get(URL)
  resposta.raise_for_status()
except HTTPError as exc:
  print(exc)
else:
  robots = resposta.text
  print(robots)
```

```python
# Verifique se a palavra  top  ou  charts  está presente no conteúdo do texto. 
# Se sim, imprima  True , senão imprima  False .

if "top" in robots or "charts" in robots:
  print("True")
else:
  print("False")
```

```python
#GETTING PAGE CONTENT 

from bs4 import BeautifulSoup
import requests
import pandas as pd
import re


page = requests.get("https://www.imdb.com/chart/top/")

soup = BeautifulSoup(page.content, "html.parser")
#print(soup.prettify())

#SCRAPPING MOVIE NAMES
scraped_movies = soup.find_all('td', class_ = 'titleColumn')

#PARSING MOVIE NAMES
movies = []
for movie in scraped_movies :
  movie = movie.get_text().replace("\n", "")
  movie = re.sub(' +', ' ', movie)
  movie = movie.strip(" ")
  movies.append(movie)

#SPLITTING POSITION AND NAME
position_ = []
movies_ = []
for m in movies:
  position = m.split(".")[0]
  position_.append(position)
  movies = m.split(".")[1]
  movies_.append(movies)

#SEPARATING NAME AND RELEASED YEAR

released_year = []
movie_names = []
error = []

for movie in movies_:
  try:    
    movie_names.append(re.split("\(|\)", movie)[0])
    released_year.append(re.split("\(|\)", movie)[1])  
  except:
    error.append(movie)

#GETTING THE TOP 10
p_10 =  position_[0:10]
m_10 = movie_names[0:10]
y_10 =  released_year[0:10]
```

```python
#SCRAPPING FILM RATING
scraped_rates = soup.find_all('td', class_ = 'ratingColumn imdbRating')
rates = []
for rate in scraped_rates:
  rate = rate.get_text().replace("\n","")
  rates.append(rate)
r_10 = rates[0:10]  
```

```python
# STORING SCRAPED DATA
import pandas as pd
data = pd.DataFrame()
data['Positions'] = p_10
data['Movies'] = m_10
data['Released_Year'] = y_10
data['Qualification'] = r_10

data
```

```python
#SAVING IN A CSV FILE
data.to_csv("Top_10_movies.csv")
```
